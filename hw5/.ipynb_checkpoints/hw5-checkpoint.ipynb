# I: Growth of Matter Perturbations - Matter and Radiation

The equation for the growth of a matter perturbation in linear theory above the Jean‚Äôs scale is given by:
\begin{align*}
    \ddot{\delta}_m + 2H\dot{\delta}_m &= 4\pi G\bar{\rho}_m\delta_m.
\end{align*}
In what follows, we will assume that the Universe is only made of matter and radiation, and is flat. Furthermore, we will ignore any fluctuations in the radiation, although it will contribute to the expansion rate ($H$).

- Show that this above evolution equation can be written as:
\begin{align*}
    \frac{d}{da}\left(a^3H\frac{d\delta_m}{da}\right) &= 4\pi G\bar{\rho}_{m,0}\frac{\delta_m}{Ha^2}
\end{align*}

### *Solution:*

We'll show these two equations are equivalent starting from the scale factor version (note I will assume $\dot{a}$ is constant under $d/da$):
\begin{align*}
    \frac{d}{da}\left(a^3H\frac{d\delta_m}{da}\right) &= 4\pi G\bar{\rho}_{m,0}\frac{\delta_m}{Ha^2}, \\
    \frac{d}{da}\left(a^2\dot{a}\frac{d\delta_m}{da}\right) &= 4\pi G\bar{\rho}_{m}a^2\frac{\delta_m}{\dot{a}}, \\
    \frac{\dot{a}}{a^2}\left(a^2\dot{a}\frac{d^2\delta_m}{da^2} + 2a\dot{a}\frac{d\delta_m}{da}\right) &= 4\pi G\bar{\rho}_{m}\delta_m, \\
    \dot{a}^2\frac{d^2\delta_m}{da^2} + 2\frac{\dot{a}^2}{a}\frac{d\delta_m}{da} &= 4\pi G\bar{\rho}_{m}\delta_m.
\end{align*}
From the chain rule:
\begin{align*}
    \frac{d}{da} &= \frac{dt}{da}\frac{d}{dt} \\
    &= \frac{1}{\dot{a}}\frac{d}{dt}.
\end{align*}
Applying this:
\begin{align*}
    \frac{d^2\delta_m}{dt^2} + 2\frac{\dot{a}}{a}\frac{d\delta_m}{dt} &= 4\pi G\bar{\rho}_{m}\delta_m, \\
    \ddot{\delta}_m + 2H\dot{\delta}_m &= 4\pi G\bar{\rho}_{m}\delta_m.
\end{align*}

- Define $y = a/a_{\text{eq}}$, where $ùëé_{\text{eq}}$ is the scale factor at matter-radiation equality. Show that the Hubble parameter can be written as:
\begin{align*}
    H(y) &= \frac{A}{y^2}\sqrt{1 + y}.
\end{align*}
Work out $A$ in terms of physical constants and the energy density at equality.

### *Solution:*

We'll start from the first Friedmann equation for matter and radiation and reference to $a_{\text{eq}}$:
\begin{align*}
    H^2 &= \frac{8\pi G}{3}\left(\bar{\rho}_m + \bar{\rho}_r\right) \\
    &= \frac{8\pi G}{3}\left(\bar{\rho}_{m,\text{eq}}\left(\frac{a}{a_{\text{eq}}}\right)^{-3} + \bar{\rho}_{r,\text{eq}}\left(\frac{a}{a_{\text{eq}}}\right)^{-4}\right) \\
    &= \frac{8\pi G}{3}\left(\bar{\rho}_{m,\text{eq}}y^{-3} + \bar{\rho}_{r,\text{eq}}y^{-4}\right) \\
\end{align*}
At $a_{\text{eq}}$, we have $\bar{\rho}_{m,\text{eq}} = \bar{\rho}_{r,\text{eq}} = \bar{\rho}_{\text{eq}}/2$:
\begin{align*}
    H^2 &= \frac{4\pi G\bar{\rho}_{\text{eq}}}{3}\left(y^{-3} + y^{-4}\right) \\
    &= \frac{4\pi G\bar{\rho}_{\text{eq}}}{3y^4}\left(1 + y\right), \\
    H &= \frac{\sqrt{4\pi G\bar{\rho}_{\text{eq}}}}{\sqrt{3}y^2}\sqrt{1 + y} \\
    &= \frac{A}{y^2}\sqrt{1 + y},
\end{align*}
where:
\begin{align*}
    A &= \sqrt{\frac{4\pi G\bar{\rho}_{\text{eq}}}{3}}.
\end{align*}

- Substitute this into the equation for growth and show that it can be written as:
\begin{align*}
    \frac{d^2\delta_m}{dy^2} + \frac{2 + 3y}{2y(1+y)}\frac{d\delta_m}{dy} &= \frac{3}{2}\frac{\delta_m}{y(1+y)}.
\end{align*}

### *Solution:*

We'll make the following substitutions in this part:
\begin{align*}
    a &= ya_{\text{eq}}, \\
    \frac{d}{da} &= \frac{1}{a_{\text{eq}}}\frac{d}{dy}, \\
    H &= \frac{A}{y^2}\sqrt{1 + y}, \\
    \bar{\rho}_{m,0} &= \frac{1}{2}\bar{\rho}_{\text{eq}}a_{\text{eq}}^3, \\
    4\pi G\bar{\rho}_{\text{eq}} &= 3A^2,
\end{align*}
Starting now from the growth equation:

\begin{align*}
    \frac{d}{da}\left(a^3H\frac{d\delta_m}{da}\right) &= 4\pi G\bar{\rho}_{m,0}\frac{\delta_m}{Ha^2}, \\
    \frac{1}{a_{\text{eq}}}\frac{d}{dy}\left(y^3a_{\text{eq}}^3\frac{A}{y^2}\sqrt{1+y}\frac{1}{a_{\text{eq}}}\frac{d\delta_m}{dy}\right) &= 2\pi G\bar{\rho}_{\text{eq}}a_{\text{eq}}^3\frac{y^2\delta_m}{Ay^2\sqrt{1+y}a_{\text{eq}}^2}, \\
    Aa_{\text{eq}}\frac{d}{dy}\left(y\sqrt{1+y}\frac{d\delta_m}{dy}\right) &= \frac{3}{2}A^2a_{\text{eq}}\frac{\delta_m}{A\sqrt{1+y}}, \\
    \frac{d}{dy}\left(y\sqrt{1+y}\frac{d\delta_m}{dy}\right) &= \frac{3}{2}\frac{\delta_m}{\sqrt{1+y}}, \\
    y\sqrt{1+y}\frac{d^2\delta_m}{dy^2} + \left(\sqrt{1 + y} + \frac{y}{2\sqrt{1 + y}}\right)\frac{d\delta_m}{dy} &= \frac{3}{2}\frac{\delta_m}{\sqrt{1+y}}, \\
    \frac{d^2\delta_m}{dy^2} + \left(\frac{1}{y} + \frac{1}{2(1 + y)}\right)\frac{d\delta_m}{dy} &= \frac{3}{2}\frac{\delta_m}{y(1+y)}, \\
    \frac{d^2\delta_m}{dy^2} + \frac{2 + 3y}{2y(1+y)}\frac{d\delta_m}{dy} &= \frac{3}{2}\frac{\delta_m}{y(1+y)}.
\end{align*}

- Verify (by eg. direct substitution, you may use Mathematica) that the solutions to this
equation are:
\begin{align*}
    \delta_m \propto 1 + \frac{3}{2}y,
\end{align*}
and:
\begin{align*}
    \delta_m \propto \left(1 + \frac{3}{2}y\right)\ln\left(\frac{\sqrt{1+y} + 1}{\sqrt{1+y} - 1}\right) - 3\sqrt{1 + y}.
\end{align*}

### *Solution:*

Checking the first solution by hand:
\begin{align*}
    \frac{d^2\delta_m}{dy^2} + \frac{2 + 3y}{2y(1+y)}\frac{d\delta_m}{dy} - \frac{3}{2}\frac{\delta_m}{y(1+y)} &= \frac{2 + 3y}{2y(1+y)}\frac{3}{2} - \frac{3}{2}\frac{1}{y(1+y)}\left(1 + \frac{3}{2}y\right) \\
    &= \frac{3}{2}\left(\frac{2 + 3y}{2y(1+y)} - \frac{2 + 3y}{2y(1+y)}\right) \\
    &= 0.
\end{align*}
Now let's get the derivatives of the second solution:

```python
y = sympy.symbols('y')
delta_m = (1 + 3*y/2)*sympy.ln((sympy.sqrt(1 + y) + 1)/(sympy.sqrt(1 + y) - 1)) - 3*sympy.sqrt(1 + y)
delta_m
```

```python
dm_dy = sympy.simplify(sympy.diff(delta_m, y))
dm_dy
```

```python
d2m_dy2 = sympy.simplify(sympy.diff(dm_dy, y))
d2m_dy2
```

We'll use these to verify the second solution:

```python
solution = d2m_dy2 + (2 + 3*y)/(2*y*(1+y))*dm_dy - 3/(2*y*(1+y))*delta_m
solution
```

```python
sympy.simplify(solution)
```

Therefore these are the two solutions.

- Check to see that the growing modes at early and late times match your expectation (for radiation and matter domination respectively).

### *Solution:*

For matter domination, we have $H = 2/(3t)$, therefore our growth equation simplifes to:
\begin{align*}
    \ddot{\delta_m} + \frac{4}{3t}\dot{\delta_m} - \frac{2}{3t^2}\delta_m = 0,
\end{align*}
which has the growing solution $\delta_m \propto a$. This is our expectation for matter domination. Our more general growing mode solution is:
\begin{align*}
    \delta_m \propto 1 + \frac{3a}{2a_{\text{eq}}}.
\end{align*}
But at late times during matter domination, we have $a \gg a_{\text{eq}}$, therefore:
\begin{align*}
    \delta_m &\propto 1 + \frac{3a}{2a_{\text{eq}}} \approx \frac{3a}{2a_{\text{eq}}}, \\
    \to \delta_m &\propto a,
\end{align*}
so this matches our expectations for matter domination. For radiation domination, the growth equation simplifies to:
\begin{align*}
    \ddot{\delta_m} + 2H\dot{\delta_m} &= 0,
\end{align*}
which has the growing solution $\delta_m \propto \ln(a)$, setting our expectation for radiation domination. This does not directly match our growing mode, however over a short timescale, $\ln(a)$ is effectively constant, and we also note that during early times in radiation domination, $a \ll a_{eq}$, so:
\begin{align*}
    \delta_m \propto 1 + \frac{3a}{2a_{\text{eq}}} \approx 1,
\end{align*}
which is also effectively constant. Therefore our general result matches our expectation for radiation domination.

# II: Spherical Collapse

In class, we used the following parametric solution for a spherical overdensity of mass $M$ and energy $E$:
\begin{align*}
    r(\theta) = A(1 - \cos\theta),
\end{align*}
and:
\begin{align*}
    t(\theta) = B(\theta - \sin\theta),
\end{align*}
with:
\begin{align*}
    A &= \frac{GM}{2|E|},
\end{align*}
and $A^3 = GMB^2$. Verify (either by solving or substituting) this solution.

### *Solution:*

The equations for spherical collapse are given as:
\begin{align*}
    \ddot{r} = -\frac{GM}{r^2}, \\
    \frac{1}{2}\dot{r}^2 - \frac{GM}{r} = E.
\end{align*}
We will verify the solutions by substitution. First lets find derivatives using the chain rule:
\begin{align*}
    \frac{d}{d\theta} &= \frac{dt}{d\theta}\frac{d}{dt} \\
    &= t'\frac{d}{dt}, \\
    \to \frac{d}{dt} &= \frac{1}{t'}\frac{d}{d\theta} \\
    &= \frac{1}{B(1 - \cos\theta)}\frac{d}{d\theta}, \\
    \frac{d^2}{d\theta^2} &= \frac{d}{d\theta}\left(t'\frac{d}{dt}\right) \\
    &= t''\frac{d}{dt} + \left(t'\right)^2\frac{d^2}{dt^2}, \\
    \to \frac{d^2}{dt^2} &= \frac{1}{\left(t'\right)^2}\left(\frac{d^2}{d\theta^2} - t''\frac{d}{dt}\right) \\
    &= \frac{1}{\left(t'\right)^2}\left(\frac{d^2}{d\theta^2} - \frac{t''}{t'}\frac{d}{d\theta}\right) \\
    &= \frac{1}{B^2(1 - \cos\theta)^2}\left(\frac{d^2}{d\theta^2} - \frac{\sin\theta}{(1 - \cos\theta)}\frac{d}{d\theta}\right).
\end{align*}
Now lets verify the first equation:
\begin{align*}
    \ddot{r} + \frac{GM}{r^2} &= \frac{1}{B^2(1 - \cos\theta)^2}\left(\frac{d^2}{d\theta^2} - \frac{\sin\theta}{(1 - \cos\theta)}\frac{d}{d\theta}\right)A(1 - \cos\theta) + \frac{GM}{A^2(1 - \cos\theta)^2} \\
    &= \frac{1}{(1 - \cos\theta)^2}\left[\frac{A}{B^2}\left(\cos\theta - \frac{\sin^2\theta}{(1 - \cos\theta)}\right) + \frac{GM}{A^2}\right] \\
    &= \frac{1}{(1 - \cos\theta)^2}\left[\frac{A}{B^2}\left(\cos\theta - \frac{1 - \cos^2\theta}{(1 - \cos\theta)}\right) + \frac{GM}{A^2}\right] \\
    &= \frac{1}{(1 - \cos\theta)^2}\left[\frac{GM}{A^2} - \frac{A}{B^2}\right] \\
    &= \frac{1}{(1 - \cos\theta)^2}\left[\frac{GM}{A^2} - \frac{GM}{A^2}\right] \\
    &= 0.
\end{align*}
Verifying the second equation:
\begin{align*}
    \frac{1}{2}\dot{r}^2 - \frac{GM}{r} - E &= \frac{1}{2}\frac{1}{B^2(1 - \cos\theta)^2}\left(\frac{d}{d\theta}A(1 - \cos\theta)\right)^2 - \frac{GM}{A(1 - \cos\theta)} - E \\
    &= \frac{A^2\sin^2\theta}{2B^2(1 - \cos\theta)^2} - \frac{GM}{A(1 - \cos\theta)} - E \\
    &= \frac{GM}{A(1 - \cos\theta)}\left(\frac{\sin^2\theta}{2(1 - \cos\theta)} - 1\right) - E \\
    &= \frac{GM}{A(1 - \cos\theta)}\left(\frac{1}{2}(1 + \cos\theta) - 1\right) - E \\
    &= -|E| - E \\
    &= E - E \\
    &= 0.
\end{align*}
In the last step, we recognize since $E < 0$ for spherical collapse, $|E| = -E$.

# III: Equality Scale

As we saw in class, the equality scale sets the turnover in the matter power spectrum. Show that the equality scale is given by:
\begin{align*}
    k_{\text{eq}} &= a_{\text{eq}}H_{\text{eq}} = H_0\sqrt{\frac{2\Omega_{m,0}}{a_{\text{eq}}}}.
\end{align*}
Estimate this scale for a cosmology roughly like the one we live in.

### *Solution:*

The Hubble radius is $H^{-1}$ or $(aH)^{-1}$ in comoving coordinates, and the wavenumber corresponding to the horizon is therefore the inverse of this:
\begin{align*}
    k_{\text{eq}} &= a_{\text{eq}}H_{\text{eq}}.
\end{align*}
For matter domination, the hubble parameters can be related as:
\begin{align*}
    H^2 = H_0^2\Omega_{m,0}a^{-3} &= H_{\text{eq}}^2\Omega_{m,\text{eq}}\left(\frac{a}{a_{\text{eq}}}\right)^{-3}, \\
    H_{\text{eq}} &= H_0\sqrt{\frac{\Omega_{m,0}a^{-3}a^3}{\Omega_{m,\text{eq}}a_{\text{eq}}^3}} \\
    &= H_0\sqrt{\frac{2\Omega_{m,0}}{a_{\text{eq}}^3}},
\end{align*}
so:
\begin{align*}
    k_{\text{eq}} &= a_{\text{eq}}H_{\text{eq}} \\
    &= \frac{H_0}{c}\sqrt{\frac{2\Omega_{m,0}}{a_{\text{eq}}}},
\end{align*}
where we insert $c$ at the end to get the correct units. For a cosmology roughly like the one we have, we will use $H_0 = 70h$ km/s/Mpc, $\Omega_{m,0} = 0.3$, and $a_{\text{eq}} = 1/(1+z_{\text{eq}}) \approx 1/3400$. Given these values, we estimate the scale of $k_{\text{eq}}$ to be approximately:

```python
H0 = 70 * u.km/u.s/u.Mpc
c = 3e5 * u.km/u.s
Omega_m = 0.3
a_eq = 1/3400
k_eq = (H0/c*np.sqrt(2*Omega_m/a_eq))
k_eq
```

- Compute the transfer function $T(k)$ from the power spectrum and plot it. Normalize it to 1 at large scales. Compare the shape to what we calculated in class.

### *Solution:*

The power spectrum at time $t$ can be written as:
\begin{align*}
    P(k,t) = T^2(k)\frac{D_+^2(t)}{D_+(t_i)}P(k,t_i).
\end{align*}
The initial power spectrum has the power law form:
\begin{align*}
    P(k,t_i) = Ak^{n_s}.
\end{align*}
Therefore the transfer function can be written as:
\begin{align*}
    T(k) &\propto \sqrt{\frac{P(k,t)}{P(k,t_i)}} \\
    &\propto \sqrt{\frac{P(k,t)}{k^{n_s}}},
\end{align*}
which we will normalize to 1 at small $k$:

```python
Tk_linear = np.sqrt(Pk_linear/k_linear**n_s)
Tk_linear /= Tk_linear[0]

fig, ax = plt.subplots(figsize=(5,4))#,dpi=300)
ax.plot(np.log10(k_linear), np.log10(Tk_linear), '-k')
ax.set_ylabel(r'$\log_{10}T(k)$')
ax.set_xlabel(r'$\log_{10}k$ (Mpc$^{-1}$)')
ax.set_title(r'Linear theory Transfer function ($z = 0$)')
fig.tight_layout()
```

This has the expected shape of being flat at small $k$ and dropping off at large $k$ due to suppression of subhorizon modes during radiation domination.

- Calculate $\sigma_8$ from the linear power spectrum. Compare this to the value given above.

### *Solution:*

Recall: $\sigma_8^2$ is defined as the variance calculated using a top hat window function of radius 8 Mpc/$h$:
\begin{align*}
    \sigma_8^2 &= \int d\ln k\Delta_8^2(k) \\
    &= \int d\ln kW_8^2(k)\Delta^2(k) \\
    &= \int d\ln k\frac{9}{(kR_8)^6}\left(\sin(kR_8) - kR_8\cos(kR_8)\right)^2\frac{k^3P(k)}{2\pi^2} \\
    &= \frac{9}{2\pi^2}\int
    d\ln k\left(\sin(kR_8) - kR_8\cos(kR_8)\right)^2\frac{k^3}{(kR_8)^6}P(k) \\
    &= \frac{9}{2\pi^2R_8^6}\int
    d\ln k\left(\sin(kR_8) - kR_8\cos(kR_8)\right)^2k^{-3}P(k).
\end{align*}
Doigna. quick change of integration variables, since I'd prefer to work with $\log_{10}k$:
\begin{align*}
    \ln k &= \ln(10)\log_{10}k, \\
    d\ln k &= \ln(10)d\log_{10}k,
\end{align*}
so:
\begin{align*}
    \sigma_8^2 &= \frac{9\ln(10)}{2\pi^2R_8^6}\int
    d\log_{10}k\left(\sin(kR_8) - kR_8\cos(kR_8)\right)^2k^{-3}P(k).
\end{align*}
No need to simplify more than this since we can do the rest numerically. We'll use `scipy.interpolate` to make the function conversions and integration easier:

```python
h = 0.6711
R8 = 8/h # Mpc

# interpolated
log10_Pk = interp.interp1d(np.log10(k_linear), np.log10(Pk_linear), kind='cubic')

def integrand(log10k):
    k = 10**log10k
    return (np.sin(k*R8) - k*R8*np.cos(k*R8))**2*k**(-3)*10**(log10_Pk(log10k))

# integrate
var = 9*np.log(10)/(2*np.pi**2*R8**6)*spi.quad(integrand, np.log10(k_linear)[0], np.log10(k_linear)[-1])[0]
sigma_8 = np.sqrt(var)
sigma_8
```

This is not quite the value $\sigma_8 = 0.834$ we were given, but it's at least the right order of magnitude.

- Now plot the ratio of the non-linear to linear power spectrum. Explain the shape of this ratio on large scales based on what you know about linear structure growth. At what $k$ does linear structure growth start to break down (should be redshift dependent)?

### *Solution:*

So we are dealing with the ratio:
\begin{align*}
    f &= \frac{P_{\text{nl}}(k,z)}{P_{\text{lin}}(k,z=0)},
\end{align*}
But this doesn't really make sense for the $z \ne 0$ cases since they are each scaled by an overall constant due to the growth factor. We'll fix this by normalizing each $P_{\text{nl}}(k,z)$ such that $P_{\text{nl}}(k=10^{-2}\text{ Mpc}^{-1},z) = P_{\text{lin}}(k=10^{-2}\text{ Mpc}^{-1},z=0)$, which is valid since the linear and nonlinear spectra are practically the same on those scales.

```python
fig, ax = plt.subplots(figsize=(5,4))#,dpi=300)
ax.axhline(0, color='k', linestyle='dashed')
ax.axhline(np.log10(2), color='r', linestyle='dashed')
for z in z_list:
    log10f = np.log10(Pk_nonlinear[z]) - log10_Pk(np.log10(k_nonlinear[z]))
    ax.plot(np.log10(k_nonlinear[z]), log10f - log10f[0],
            label=f'nonlinear (z={z})')
ax.set_ylabel(r'$\log_{10}(\frac{P_{nl}}{P_{lin}})(k)$')
ax.set_xlabel(r'$\log_{10}k$ (Mpc$^{-1}$)')
ax.set_title(r'Linear + Nonlinear theory $P(k)$ ratio (normalized)')
ax.legend()
fig.tight_layout()
```

Here I'm using $f = 2$ as our criterion for linear growth starting to break down. At large scales, this ratio is approximately flat, i.e. linear growth is a good approximation. At $z=2$, the linear theory starts to break down above $k \ge 1$ Mpc, whereas at $z = 0$ it start to break down above $k \ge 1/3$ Mpc. Overall, linear growth becomes a worse approximation for large $k$ and for smaller $z$.

- From this plot, extract the growth function as a function of redshift. How are you choosing to normalize this?

### *Solution:*

The growth function and the power spectrum are related as:
\begin{align*}
    \delta(\vec{k},t) &= \frac{D(t)}{D(t_i)}\delta(\vec{k},t_i),
\end{align*}
or:
\begin{align*}
    \delta(\vec{k},z) &= \frac{D(z)}{D(z=0)}\delta(\vec{k},z=0).
\end{align*}
Let's normalize as $D(z=0) = 1$, so:
\begin{align*}
    \delta(\vec{k},z) &= D(z)\delta(\vec{k},z=0), \\
    D(z) &= \frac{\delta(\vec{k},z)}{\delta(\vec{k},z=0)}.
\end{align*}
This is exactly the ratio we were calculating last time! But now we care about the constant offsets. We'll calculate these using the lowest $k$ mode:

- Calculate the growth function for the cosmology of the simulation, using the method below. Compare this to your extracted growth function. How well do they agree?

## Computing the Growth Function

Start by defining $g(a)$ as:
\begin{align*}
    D(a) \equiv g(a)a,
\end{align*}
where $D(a)$ is the linear growth function. Note that for a matter dominated Universe, $D(a) = a$, and so $g(a) = 1$. Define primes as derivatives with respect to $\ln a$. Then you can show (but you do not have to) that $g(a)$ satisfies the following equation (eg. Linder 2005):
\begin{align*}
    g'' + \left(\frac{5}{2} - \frac{3}{2}w(a)\Omega_{\text{DE}}(a)\right)g' - \frac{3}{2}(1 - w(a))\Omega_{\text{DE}}(a)g &= 0.
\end{align*}
where $w(a)$ is the equation of state of dark energy, and $\Omega_{\text{DE}}(a)$ is the fraction of the total energy density in dark energy. Note that the quantity that appears in the equation is $\Omega_{\text{DE}}(a)$, i.e. the redshift dependent quantity.

This equation is quite general, and works for many different cosmologies with varying dark energy. Note that many of the other approximations you may find in the literature do not.

For the cosmology of the simulation, $w(a) = -1$, so the equation simplifies considerably. Finally, remember to multiply by $a$ to get the growth rate.

In order to determine the growth rate for the previous problem, integrate this equation from some very early time (remember that you are integrating in $\ln a$); eg. $\ln a = -5.0$. At this time you may assume that the Universe is matter dominated, and so $g(a) = 1$, which you can use to set the initial conditions. Then integrate forward. When you get to $a = 1$, you should find that $g(a=1) \approx 0.78$.

### *Solution:*

The growth equation using $w(a) = -1$ simplifies to:
\begin{align*}
    g'' + \frac{1}{2}\left(5 + 3\Omega_{\text{DE}}(a)\right)g' - 3\Omega_{\text{DE}}(a)g &= 0.
\end{align*}
We'll need to find a form for $\Omega_{\text{DE}}(a)$. We can solve this as:
\begin{align*}
    \Omega_{\text{DE}} &= 1 - \Omega_m \\
    &= 1 - \frac{\rho_m}{\rho_c} \\
    &= 1 - \frac{\rho_{m,0}a^{-3}}{\rho_{c,0}}\frac{\rho_{c,0}}{\rho_c} \\
    &= 1 - \Omega_{m,0}a^{-3}\frac{H_0^2}{H^2} \\
    &= 1 - \frac{\Omega_{m,0}a^{-3}}{\Omega_{m,0}a^{-3} + \Omega_{\text{DE},0}}.
\end{align*}
We will use $\Omega_{m,0} = 0.3175$ and $\Omega_{\text{DE},0} = 0.6825$. Let's check that our solution for $\Omega_{\rm{DE}}$ behaves reasonably:

```python
Omega_m0 = 0.3175
Omega_DE0 = 1 - Omega_m0

def Omega_DE(a):
    return 1 - Omega_m0*a**(-3)/(Omega_m0*a**(-3) + Omega_DE0)
```

```python
a = np.linspace(1e-5,10,int(1e6))

plt.plot(a, Omega_DE(a))
plt.semilogx()
plt.xlabel('a')
plt.ylabel(r'$\Omega_{\rm{DE}}$')
```

Great! Now we can integrate. Note we have the initial conditions $g(a) = 1$ and $dg/da = dg/\log_{10}a = 0$:

```python
# Define the differential equation system
def growth(log10_a, y):
    
    a = 10**log10_a
    g, g_prime = y
    
    dg_da = g_prime
    d2g_da2 = -0.5*(5 + 3*Omega_DE(a))*g_prime - 3*Omega_DE(a)*g
    return [dg_da, d2g_da2]

# initial conditions
g0 = 1
g_prime0 = 0
log10_a = np.linspace(-5,0,1000)

# solve
solution = spi.solve_ivp(growth, [log10_a[0],log10_a[-1]], [g0, g_prime0], t_eval=log10_a, method='RK45')
g = solution.y[0]
log10_D = np.log10(g) + log10_a
print(g[-1])
```

This is a bit higher than what we should have gotten. Finally, use:
\begin{align*}
    1 + z &= \frac{1}{a}, \\
    \log_{10}z &= \log_{10}\left(\frac{1}{a} - 1\right), \\
\end{align*}

```python
log10_z = np.log10(1/10**log10_a - 1)

fig, ax = plt.subplots(figsize=(5,4))#,dpi=300)
# make sure it's normalized
ax.plot(np.log10(np.array(z_list[1:])), log10D[1:], 'o')
ax.plot(log10_z[:-1], log10_D[:-1], '-k')
ax.set_ylabel(r'$\log_{10}D$')
ax.set_xlabel(r'$\log_{10}z$')
ax.set_title(r'Growth factor')
fig.tight_layout()
```

They are not an exact match, but they are close and the qualitative behavior of the integrated curve is as expected.

- Using the linear power spectrum, calculate $\sigma(M)$, where $M = (4/3)\pi R^3\bar{\rho}_m$. Plot this quantity.

We can calculate this as:
\begin{align*}
    \sigma^2(M) &= \int_0^k\frac{k^2}{2\pi^2}P(k)W^2(kR)dk.
\end{align*}
We haven't been given a window function, so I'll use a Gaussian:
\begin{align*}
    W(kR) &= \exp\left(-\frac{(kR)^2}{2}\right).
\end{align*}
We'll use our upper and lower range of halo masses to define our integration bounds, and use $\bar{\rho}_m = M_{\text{tot}}/V$.

```python
def sigma_sq_integrand(k, R):
    index = np.argmin(np.abs(k_linear - k))
    W = np.exp(-(k*R)**2/2)
    return k**2/(2*np.pi**2)*Pk_linear[index]*W**2

N_M = 500
h = 0.6711
L = h*u.Gpc.to(u.Mpc)
V = L**3
M = np.linspace(np.min(M_200b), np.max(M_200c), N_M)
rho = Mtot/V
R = M/(rho*4/3*np.pi)

sigma = np.zeros(N_M)
for i in range(N_M):
    if i == 0:
        continue
    sigma[i] = np.sqrt(spi.quad(sigma_sq_integrand, k_linear[0], k_linear[-1], args=(R[i])))[0]
```
